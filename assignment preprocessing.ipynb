```python
# Impute missing values in the 'Age' columnn with the mean
titanic = ships
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
# Impute missing values in the 'Cabin' columnnwith the constant value
titanic.Cabin = titanic.Cabin.fillna(0)
# Impute missing values in the 'Embarked' columnnwith the 'S'
titanic.Embarked = titanic.Embarked.fillna('S')
```

```python
titanic.info
```

```python
import seaborn as sns

# create seaborn boxplot for target variable
sns.boxplot(x='Age', data=titanic)
```

```python
#  Z Score method
data = titanic['Age']

# calculate the z score
z_scores = (data - data.mean()) / data.std()
print("Z-Score: ",z_scores.mean())
# define a threshold for identifying outliers

threshold = 2

# normal range: (outliers) -2------ 2(outliers)

# Identify outliers
outliers = titanic[abs(z_scores) > threshold]

# print the outliers
print("outliers:")
outliers
```

```python
import seaborn as sns

# create seaborn boxplot for target variable
sns.boxplot(x='Fare', data=titanic)
```

```python
# Calculate the IQR
data = titanic["Fare"]
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for identifying outliers (e.g., 1.5 times the IQR)
threshold = 1.5

# Identify the outliers
outliers = titanic[(data < Q1 - threshold * IQR) | (data > Q3 + threshold * IQR)]

# Print the outliers
print("Outliers:")
outliers
```

```python
sns.boxplot(x='Pclass', data=titanic)
```

```python
# Calculate the IQR
data = titanic["Pclass"]
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for identifying outliers (e.g., 1.5 times the IQR)
threshold = 1.5

# Identify the outliers
outliers = titanic[(data < Q1 - threshold * IQR) | (data > Q3 + threshold * IQR)]

# Print the outliers
print("Outliers:")
outliers
```
